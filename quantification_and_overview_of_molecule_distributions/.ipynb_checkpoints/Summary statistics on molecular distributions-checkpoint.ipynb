```python
import numpy as np
import pandas as pd 
import re
from tqdm import tqdm
from glob import glob
import seaborn as sns
import gzip 
import matplotlib.pyplot as plt
from collections import defaultdict

from collections import Counter
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings('ignore')
```

```python
chrom_alias = pd.read_csv('../../fig2_ctcf_and_tar1/all_diploid_alias.txt',
                          sep="\t",
                          header=None,
                          names=['id','chrom','contig']
                         )

id_to_chr = dict(zip(chrom_alias['id'], chrom_alias['chrom']))

col_names = ['chrom',
         'centering_position',
         'strand',
         'subset_sequence',
         'reference_start',
         'reference_end',
         'query_name',
         'RG',
         'HP',
         'centered_query_start',
         'centered_query_end',
         'query_length',
         '5mC_pos',
         '5mC_score',
         'query_sequence']





def normalize_chrom_assignment(df: pd.DataFrame,
                               chrom_assignment_col: str = "chrom_assignment",
                               pos_col: str = "centering_position") -> pd.DataFrame:
    """
    Cleans chrom_assignment by:
      1. Removing underscores and everything after them.
      2. Keeping canonical chr1–22, X, Y, M (including double digits).
      3. Assigning p/q arm based on suffix or centering_position.
    """

    def clean_and_assign(s, pos):
        if not isinstance(s, str):
            return None

        # strip everything after underscore (e.g., _alt, _random)
        base = re.sub(r'_.*', '', s)

        # make sure standard chrom name
        m = re.match(r'^(chr(?:1[0-9]|2[0-2]|[1-9]|X|Y|M))', base)
        if not m:
            return None
        chrom_base = m.group(1)

        # Determine arm
        if base.endswith("p") or base.endswith("q"):
            arm = base[-1]
        else:
            try:
                # if arm on far right of contig, assign q (HPRC contigs)
                # else assign p because HPRC contifg should oriented
                arm = "q" if float(pos) > 100_000 else "p" 
            except (TypeError, ValueError):
                arm = None

        return f"{chrom_base}{arm}" if arm else chrom_base

    df[chrom_assignment_col] = [
        clean_and_assign(s, p) for s, p in zip(df[chrom_assignment_col], df[pos_col])
    ]
    return df


def get_tsvs_and_labels(file_pattern, fiberseq=False):
    """
    Reads TSV files matching the file pattern, processes chromosome assignments, 
    and returns a list of dataframes and a list of sample labels.
    """
    tsvs = []
    labels = []

    for file in tqdm(glob(file_pattern)):
        try:
            sample_tsv = pd.read_csv(file, sep="\t", index_col=False)
            # map sample contig IDs to base chromosome names (e.g., "chr1")
            sample_tsv['chr_base'] = sample_tsv['chrom'].map(id_to_chr)

            # if centering_position > 100kb → 'q', else 'p'
            threshold = 100_000
            sample_tsv['chrom_assignment'] = (
                sample_tsv['chr_base'] + np.where(sample_tsv['centering_position'].astype(int) > threshold, 'q', 'p')
            )
            if fiberseq:
                req = np.ceil(0.001 * (sample_tsv["centered_query_end"] - sample_tsv["centered_query_start"]).astype(float)).astype(int)
                req = req.clip(lower=1)  # optional: require at least 1
                # observed count from comma-separated m6a_qual
                obs = sample_tsv["m6a_qual"].fillna("").str.strip(",").str.count(",") + 1
                
                n_before=len(sample_tsv)
                sample_tsv = sample_tsv[obs >= req]
            
                print(f"Dropped {n_before - len(sample_tsv):,} rows (adaptive cutoff)")
            # optional: drop helper column
            sample_tsv.drop(columns=['chr_base'], inplace=True)
            sample_tsv = normalize_chrom_assignment(sample_tsv)
            sample_tsv = sample_tsv[sample_tsv["chrom_assignment"].notna()]

            tsvs.append(sample_tsv)
            labels.append(file.split("/")[-1].split('.')[0])
        except:
            continue
    return tsvs, labels


def plot_telomere_metrics(all_hprc_tsvs, color='steelblue', save_path=None):

    
    # process data
    telo_mol_counts = []
    telo_captured_per_sample = []
    telo_mols_per_chrom_end = []
    telo_mols_counts_by_individual = []
    telo_haps_per_end = Counter()

    for tsv in all_hprc_tsvs:
        # remove molecules that could not be assigned to a chromosome 
        tmp = tsv[["chrom", "chrom_assignment"]].dropna()
        chrom_str = tmp["chrom"].astype(str)
        
        # for swarmplot 1: total telomere molecules
        telo_mol_counts.append(len(tmp))
        
        # for swarmplot 2 and a bar chart : telomeres captured per sample and haps per end
        per_end = tmp.groupby("chrom_assignment")["chrom"].apply(
            lambda s: int(s.astype(str).str.contains("#1#", regex=False).any()) +
                      int(s.astype(str).str.contains("#2#", regex=False).any()))
        telo_captured_per_sample.append(int(per_end.sum()))
        telo_haps_per_end.update(per_end.to_dict())
        
        # for histogram telomere molecules per chrom end collect for hap1+2
        hap1 = tmp[chrom_str.str.contains("#1#", regex=False)].groupby("chrom_assignment").size()
        hap2 = tmp[chrom_str.str.contains("#2#", regex=False)].groupby("chrom_assignment").size()
        telo_mols_per_chrom_end.extend(hap1.values)
        telo_mols_per_chrom_end.extend(hap2.values)
        
        # swarmplot median counts per individual per end
        counts = (chrom_str + "|" + tmp["chrom_assignment"].astype(str)).value_counts()
        telo_mols_counts_by_individual.append(np.nanmedian(counts.values))

    # sort bar chart data
    ends, counts_bar = zip(*sorted(telo_haps_per_end.items(), key=lambda x: x[1], reverse=True))

    # plotting
    plot_styles = {
        'font.family': 'Arial',            
        'axes.labelweight': 'bold',        
        'axes.titleweight': 'bold',        
        'axes.linewidth': 0.8,             
        'xtick.major.width': 0.8,          
        'ytick.major.width': 0.8           
    }

    with plt.rc_context(plot_styles):
        fig = plt.figure(figsize=(14, 16))
        gs = gridspec.GridSpec(3, 3, height_ratios=[1, 1, 1.2], hspace=0.4, wspace=0.3)

        ax1 = fig.add_subplot(gs[0, :])
        ax1.hist(telo_mols_per_chrom_end, bins=(max(telo_mols_per_chrom_end)), color=color) # Removed hardcoded range
        ax1.set_title("Telomeres molecules per chrom end")
        ax1.set_ylabel("Frequency")
        
        med1 = np.nanmedian(telo_mols_per_chrom_end)
        ax1.axvline(med1, color='black', linestyle='--', lw=0.4)

        ax1.text(med1, 0.95, f' median={med1:.0f}', transform=ax1.get_xaxis_transform(), 
                 fontweight='bold', va='top', ha='left')

        ax2 = fig.add_subplot(gs[1, :])
        ax2.bar(range(len(ends)), counts_bar, color=color)
        ax2.set_xticks(range(len(ends)))
        ax2.set_xticklabels(ends, rotation=90, fontsize=6)
        ax2.set_ylabel("Number of haplotypes sequenced")
        ax2.set_title("Haplotypes sequenced per chromosome end")


        
        ax3 = fig.add_subplot(gs[2, 0])
        sns.swarmplot(y=telo_mol_counts, size=5, ax=ax3, color=color)
        ax3.set_xticks([]) 
        ax3.set_title("Telomere molecules\nper sample")
        ax3.set_ylabel("Count")
        
        med3 = np.nanmedian(telo_mol_counts)
        ax3.axhline(med3, color='black', linestyle='--', lw=0.4)
        ax3.text(0.55, med3, f'median={med3:.0f}', transform=ax3.get_yaxis_transform(), 
                 fontweight='bold', va='bottom', ha='left')

        ax4 = fig.add_subplot(gs[2, 1])
        sns.swarmplot(y=telo_captured_per_sample, size=5, ax=ax4, color=color)
        ax4.set_xticks([]) 
        ax4.set_title("Telomeres captured\nper sample")
        ax4.set_ylabel("") 
        
        med4 = np.nanmedian(telo_captured_per_sample)
        ax4.axhline(med4, color='black', linestyle='--', lw=0.4)
        ax4.text(0.55, med4, f'median={med4:.0f}', transform=ax4.get_yaxis_transform(), 
                 fontweight='bold', va='bottom', ha='left')

        ax5 = fig.add_subplot(gs[2, 2])
        sns.swarmplot(y=telo_mols_counts_by_individual, size=5, ax=ax5, color=color)
        ax5.set_xticks([]) 
        ax5.set_title("Median Telomere molecules\nper chrom per individual")
        ax5.set_ylabel("")
        
        med5 = np.nanmedian(telo_mols_counts_by_individual)
        ax5.axhline(med5, color='black', linestyle='--', lw=0.4)
        ax5.text(0.55, med5, f'median={med5:.1f}', transform=ax5.get_yaxis_transform(), 
                 fontweight='bold', va='bottom', ha='left')


        if save_path:
            plt.savefig(save_path, dpi=400, bbox_inches='tight')
            print(f"Figure saved to: {save_path}")
            
        plt.show()
```
